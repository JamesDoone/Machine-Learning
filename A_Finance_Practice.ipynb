# Calculating Risk

```python
tickers = ['TSLA', 'F']

sec_data = pd.DataFrame()

for i in tickers:
    sec_data[i] = wb.DataReader(i, 'google', start='2016-1-1')['Close']
```

```python
sec_data.head()
```

```python
def log_return_securities(ticker):
    return np.log((ticker) / (ticker.shift(1)))

def daily_log_return(ticker, wrkdays):
    daily_return = log_return_securities(ticker).mean() * wrkdays
    print(str(round(daily_return, 5) * 100) + '%')
```

```python
daily_log_return(sec_data,250)
```

# Risk  function

```python
def risk(ticker):
    daily_return = np.log((ticker) / (ticker.shift(1)))
    return daily_return.std() * 250**0.5
    print(str(round(daily_return.std() * 250**0.5, 5)))
```

```python
risk(sec_data)
```
